# Chapter 1 (Introduction)

## What is Statistics?

Statistics is a science that helps us design experiments, collect data, summarize it, and make inferences and decisions in the presence of uncertainty.

## Data collection can be done through different study designs
- **Retrospective studies** (using historical data),
- **Observational studies:** (monitoring real-time processes)
- **designed studies** (where researchers control variables to establish cause-and-effect relationships).

## Examples of Statistical Applications:

- **Ohm’s Law Experiment:** In this example, a student measures the relationship between voltage, resistance, and current in an electrical circuit. The population is a hypothetical set of all possible current measurements, and the sample consists of observed measurements.

- **Cell Count Comparison:** In a laboratory experiment, automated and manual counts of cells are compared to assess the accuracy of automated methods. Statistical analysis, like confidence intervals, is used to make inferences about the accuracy of the methods.

- **Rana Pipiens Muscle Study:** This experiment involves modeling the force exerted by muscle as it is stretched. The goal is to establish a statistical model that describes the relationship between muscle length and force, helping to identify how various factors affect the muscle's performance.

- **Polio Vaccine Trial:** This famous study assessed the effectiveness of the Salk polio vaccine by comparing outcomes (paralytic polio) in two groups: those who received the vaccine and those who received a placebo. Randomization ensured that the sample was representative and that the results could be generalized to the larger population.

## Definitions

- Attribute: Property of interest (quantitative or qualitative).
- Population: The entire group being studied.
- Sample: A subset of the population.
- Data: Observed values recorded from the sample.
- Statistical Model: A mathematical representation of population attributes, accounting for uncertainty.
- Population Parameter: are the components in the model that quantify the population's characteristics (like means, variances, etc.)

## 2.3 Location Measures (aka Means)
**The three averages – arithmetic, geometric, and harmonic – are known as Pythagorean means.**

### **1. Arithmetic Mean (AM)**
The arithmetic mean of \( n \) numbers \( x_1, x_2, ..., x_n \) is given by:

$$
AM = \frac{x_1 + x_2 + \dots + x_n}{n}
$$

#### **Best Used For: General averaging, normally distributed data (e.g., test scores, temperatures, salaries).**

⚠️ Not good for: Skewed distributions (outliers can heavily impact AM). Rates or ratios
 
❌ Highly Sensitive to Outliers

---



### **2. Geometric Mean (GM)**
The geometric mean of \( n \) numbers \( x_1, x_2, ..., x_n \) is given by:

$$
GM = \left( \prod_{i=1}^{n} x_i \right)^{\frac{1}{n}}
$$

or equivalently:

$$
GM = \sqrt[n]{x_1 \cdot x_2 \cdots x_n}
$$

#### **Best Used For: Growth rates(compounding values over time), proportions, and percentages**

✅ Moderately resistant to outlers 

---


### **3. Harmonic Mean (HM)**
The harmonic mean of \( n \) numbers \( x_1, x_2, ..., x_n \) is given by:

$$
HM = \frac{n}{\sum_{i=1}^{n} \frac{1}{x_i}}
$$

#### **Best Used For: Ratios, rates (e.g., speed, finance), Inversely related ratios**

✅✅ Most resistant (except for small values)

---

### <div style="text-align:center;"> Jensen's Inequality of the Means </div>

#### <div style="text-align:center;"> **Harmonic Mean (HM) ≦ Geometric Mean (GM) ≦ Arithmetic Mean (AM)** </div>


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x_values = np.linspace(1, 10, 10)

# Calculate means
arithmetic_mean = np.mean(x_values)
geometric_mean = np.exp(np.mean(np.log(x_values)))  # Equivalent to nth root of product
harmonic_mean = len(x_values) / np.sum(1 / x_values)

# Create figure
plt.figure(figsize=(8, 5))

# Plot data points
plt.plot(x_values, x_values, 'bo-', label='Data Points')

# Plot means
plt.axhline(arithmetic_mean, color='r', linestyle='--', label=f'Arithmetic Mean: {arithmetic_mean:.2f}')
plt.axhline(geometric_mean, color='g', linestyle='--', label=f'Geometric Mean: {geometric_mean:.2f}')
plt.axhline(harmonic_mean, color='purple', linestyle='--', label=f'Harmonic Mean: {harmonic_mean:.2f}')

# Labels and title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Comparison of Arithmetic, Geometric, and Harmonic Means")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

### The Winsorized Mean 

This is a robust statistical measure that reduces the impact of outliers by replacing extreme values with the nearest remaining values within a specified percentile range.

#### Formula:
1. Sort the data.
2. Replace the lowest p% and highest p% of values with the closest remaining values.
3. Compute the mean of the modified dataset.
#### Use Case:
- Useful when data contains extreme outliers.
- Provides a more stable estimate of central tendency than the regular mean.
- Often used in finance, economics, and robust statistics.

## 2.4 Variability Measures

Location measures are intuitive but provide limited insight into a sample's (dataset's) characteristics. A crucial set of descriptors are dispersion measures(Variance/Standard Deviation), which indicate the spread of the data.

### **Example: Calculating/Updating Variance**

#### **Given Initial Data:**
- Data points: **\(X = \{2, 4, 6\}\)**
- Compute initial **mean** and **variance**.

#### **Step 1: Compute Initial Mean**
$$
\mu_3 = \frac{2 + 4 + 6}{3} = \frac{12}{3} = 4
$$

#### **Step 2: Compute Initial Variance**
The sample variance formula is:

$$
s_n^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu_n)^2
$$

Substituting the values:

$$
s_3^2 = \frac{(2-4)^2 + (4-4)^2 + (6-4)^2}{3}
$$

$$
s_3^2 = \frac{4 + 0 + 4}{3} = \frac{8}{3} \approx 2.67
$$

---

#### **New Data Point: \( x_4 = 8 \)**

#### **Step 3: Update the Mean**
Using the incremental mean formula:

$$
\mu_{4} = \mu_3 + \frac{x_4 - \mu_3}{4}
$$

$$
\mu_{4} = 4 + \frac{8 - 4}{4} = 4 + \frac{4}{4} = 5
$$

---

#### **Step 4: Update the Variance**
Using the incremental variance formula:

$$
s_{4}^2 = \frac{3s_3^2 + (x_4 - \mu_3)(x_4 - \mu_4)}{4}
$$

$$
s_4^2 = \frac{3(2.67) + (8 - 4)(8 - 5)}{4}
$$

$$
s_4^2 = \frac{8 + (4 \times 3)}{4}
$$

$$
s_4^2 = \frac{8 + 12}{4} = \frac{20}{4} = 5
$$

---

#### **Final Updated Values:**
- **Updated Mean**: 
$$\mu_4 = 5 $$
- **Updated Variance**: $$s_4^2 = 5$$



#### **Variability Measures in Bayesian Statistics**
Variability measures describe the spread of a dataset, providing more insight than location measures (mean, median).

#### **1. Sample Variance & Standard Deviation**
- **Variance** measures the average squared deviation from the mean:

  $$
  s^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2
  $$

  - Uses **\( n-1 \)** (Bessel’s correction) to make it an **unbiased estimator** of population variance.

- **Standard Deviation** is the square root of variance:

  $$
  s = \sqrt{s^2} = \sqrt{\frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2}
  $$

#### **2. MAD (Median Absolute Deviation)**
- **MAD0**: Mean absolute deviation from the mean.

  $$
  MAD_0 = \frac{1}{n} \sum_{i=1}^{n} |X_i - \bar{X}|
  $$

- **MAD1**: Median absolute deviation from the median.

  $$
  MAD_1 = \text{median}(|X_i - \text{median}(X_i)|)
  $$

- Often scaled (multiplied by **1.4826**) to make it comparable to standard deviation.

#### **3. Range & Interquartile Range (IQR)**
- **Range**: 

  $$
  R = X_{\max} - X_{\min}
  $$

- **Interquartile Range (IQR)**:

  $$
  IQR = Q_3 - Q_1
  $$

- A robust variance estimator:

  $$
  \hat{\sigma}^2 = \left(\frac{IQR}{1.349}\right)^2
  $$

  - Ignores 25% of extreme observations, making it resistant to outliers.

#### **4. Skewness & Kurtosis**
- **Skewness** measures asymmetry:

  $$
  \gamma_n = \frac{\mu_3}{\mu_2^{3/2}} = \frac{\mu_3}{s^3}
  $$

  - **Positive** skew: Right tail is longer, \( \text{mean} > \text{median} \).
  - **Negative** skew: Left tail is longer, \( \text{mean} < \text{median} \).

- **Kurtosis** measures tail heaviness:

  $$
  \kappa_n = \frac{\mu_4}{\mu_2^2} = \frac{\mu_4}{s^4}
  $$

  - **Platykurtic** (\(<3\)) – Flatter than normal.
  - **Leptokurtic** (\(>3\)) – More peaked, heavy tails.
  - **Excess kurtosis**:

    $$
    \kappa_n - 3
    $$

#### **5. Robust Skewness & Kurtosis**
- **Robust Skewness** using quartiles:

  $$
  \gamma_n^* = \frac{(Q_3 - M) - (M - Q_1)}{Q_3 - Q_1}
  $$

- **Robust Kurtosis** using octiles:

  $$
  \kappa_n^* = \frac{(O_7 - O_5) + (O_3 - O_1)}{O_6 - O_2}
  $$

#### **6. Coefficient of Variation (CV)**
- **CV** expresses variability relative to the mean:

  $$
  CV = \frac{s}{\bar{X}}
  $$

- Useful for comparing datasets with different units or scales.

#### **7. z-Scores (Standardization)**
- Standardizing a sample:

  $$
  z_i = \frac{X_i - \bar{X}}{s}
  $$

- Mean becomes **0**, variance becomes **1**.

#### **8. Higher-Order Moments**
- **Raw moments**:

  $$
  m_k = \frac{1}{n} \sum_{i=1}^{n} X_i^k
  $$

  - First moment \( m_1 = \bar{X} \).

- **Central moments**:

  $$
  \mu_k = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})^k
  $$

  - **Second moment**: Variance.
  - **Third moment**: Skewness.
  - **Fourth moment**: Kurtosis.

---

#### **Applications in Machine Learning**
- **Feature Engineering** – Detecting outliers and normalizing data.
- **Probability Distributions** – Understanding model assumptions.
- **Performance Metrics** – Variability in predictions.
- **Comparing Datasets** – Handling different measurement scales.

---

#### **Final Summary**
| Measure | Formula | Strengths | Weaknesses |
|---------|---------|-----------|------------|
| **Sample Variance** \( s^2 \) | $$ s^2 = \frac{1}{n - 1} \sum (X_i - \bar{X})^2 $$ | Measures spread of data | Sensitive to outliers |
| **Alternative Variance Formula** | $$ s^2 = \frac{1}{n - 1} \left( \sum X_i^2 - n\bar{X}^2 \right) $$ | Easier for hand calculations | Same as above |
| **MAD (Mean or Median Absolute Deviation)** | $$ MAD_0 = \frac{1}{n} \sum \left| X_i - \bar{X} \right| $$ $$ MAD_1 = \text{median} \left( \left| X_i - \text{median}(X) \right| \right) $$ | Less sensitive to outliers | Not as common as variance |
| **Range** \( R \) | $$ R = X_{\max} - X_{\min} $$ | Very simple | Affected by outliers |
| **Interquartile Range (IQR)** | $$ IQR = Q_3 - Q_1 $$ | Robust to outliers | Ignores 50% of data |
| **Robust Variance Estimator** | $$ \hat{\sigma}^2 = \left(\frac{IQR}{1.349}\right)^2 $$ | Good for skewed data | Less efficient for normal data |
| **Quantiles/Percentiles** | $$ \text{Value at } p \times 100\% \text{ rank} $$ | Shows distribution shape | Requires sorting data |

## 3.2 Events and Probability

The **relative frequency interpretation of probability** states that if an experiment is repeated many times, the probability of an outcome is approximated by how often it occurs. As the number of trials increases, the relative frequency stabilizes around the theoretical probability (e.g., a fair coin landing on heads about **50%** of the time). Historical experiments, like Buffon's and Pearson's coin tosses, confirm this, showing that larger sample sizes lead to probabilities closer to their expected values.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of tosses
N = 400  # Total number of coin tosses (same as Pearson's experiment)

# Simulate coin tosses (1 = heads, 0 = tails)
coin_tosses = np.random.choice([0, 1], size=N)

# Compute cumulative proportion of heads over time
cumulative_heads = np.cumsum(coin_tosses) / np.arange(1, N + 1)

# Plot results
plt.figure(figsize=(10, 5))
plt.plot(np.arange(1, N + 1), cumulative_heads, label="Estimated Probability of Heads", color='blue')
plt.axhline(y=0.5, color='r', linestyle='--', label="True Probability (0.5)")
plt.xlabel("Number of Tosses")
plt.ylabel("Relative Frequency of Heads")
plt.title("Convergence of Relative Frequency to Probability")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

**This code demonstrates that as an experiment is repeated multiple times, the relative frequency gradually stabilizes around the theoretical probability, which in this case is 50%.**

### **What if the experiments cannot be repeated? Like drug testing and 1-time events**

### In such cases we can define probability subjectively as a measure of strength of belief, Human Intutition. Among several possible ways to define probability, these three are the most common ones.

- **Frequentist Approach**: The probability of an event is based on how often it happens in repeated trials. For example, if you flip a fair coin many times, you expect heads to come up about **50%** of the time.  

- **Subjective Approach**: Probability is a personal belief about how likely something is to happen. For example, if a doctor says there's a **70% chance** of rain tomorrow, that number reflects their confidence based on available information, not repeated experiments.  

- **Classical Approach**: Probability is calculated by dividing the number of ways an event can happen by the total number of possible outcomes. For example, in rolling a fair die, the chance of rolling a **3** is **1 out of 6** because there are 6 equally likely sides.

### **Independent vs. Mutually Exclusive Events**
---
### **Independent Probabilites** = P(A) and P(B) = P(A) ∩ P(B)
- **Independent Events**: Two events are **independent** if one happening does **not** affect the probability of the other. Since they can occur **at the same time**, we multiply their probabilities:

  $$
  P(A \text{ and } B) = P(A) \times P(B)
  $$

  **Example:** Rolling a die and flipping a coin are independent. The probability of rolling a **6** and getting **heads** is:

  $$
  P(6) \times P(H) = \frac{1}{6} \times \frac{1}{2} = \frac{1}{12}
  $$
---
### **Mutualy Exclusive (Dependent Probabilites)** = P(A) or P(B) = P(A) ∪ P(B)
- **Mutually Exclusive Events**: Two events are **mutually exclusive** if they **cannot** happen at the same time. Since it's **one or the other**, we **add** their probabilities:

  $$
  P(A \text{ or } B) = P(A) + P(B)
  $$

  **Example:** Rolling a **3** or a **5** on a die is mutually exclusive (you can't roll both). The probability is:

  $$
  P(3) + P(5) = \frac{1}{6} + \frac{1}{6} = \frac{1}{3}
  $$

   ### **What is P(A ∪ B) if events A and B are not mutually exclusive?**  
    
    To determine this, we use the **inclusion-exclusion rule**:  
        $$P(A ∪ B) = P(A) + P(B) − P(A ∩ B)$$   
    
    If events **A** and **B** are mutually exclusive, then:  
        $$P(A ∩ B) = 0$$

    For more than 2 events: 
        $$P(A ∪ B ∪ C) = P(A) + P(B) + P(C) − P(A ∩ B) - P(A ∩ C) P(B ∩ C) + P(A ∩ B ∩ C)$$  
---
Thus, **independent events can occur together**, while **mutually exclusive events cannot**. 🚀

### Compliment of a Probability 
- Counterpart event called its **complement**, like the opposite of the outcome. 
For any pair of **complementary events** ( A ) and ( A^c ), the following fundamental probability rules hold:  

$$P(A) + P(A^c) = 1$$
$$P(A) = 1 - P(A^c)$$
$$P(A^c) = 1 - P(A)$$  

These equations express that the probability of an event and its complement always sum to 1.


## 3.5 Counting Principles

This is for when we have simantanueous events occuring that are **mutually exclusive**. Instead of calculating all possibilities for the large sample size, combinatorial methods like factorials, binomial coefficients (n-choose-k), and permutations help count desired outcomes efficiently.


### **Example Problem: Counting Principles (Mutually Exclusive Events)**  

A box contains **5 different colored balls**: **red, blue, green, yellow, and white**. You randomly pick **3 balls** **without replacement**. What is the probability of selecting **red, blue, and green**?  

#### **Solution (Using Counting Principles):**  
Since order **does not matter**, we use **combinations** to count the total number of ways to choose any **3 balls from 5**:

$$
\binom{5}{3} = \frac{5!}{3!(5-3)!} = \frac{5!}{3!2!} = \frac{5 \times 4}{2 \times 1} = 10
$$

There is only **1** favorable outcome (choosing red, blue, and green), so the probability is:

$$
P(\text{red, blue, green}) = \frac{1}{10}
$$

Thus, the probability of selecting **red, blue, and green** is **1/10 or 10%**.

## 3.6 Conditional Probability and Independence of Events

A conditional probability is the probability of one event if we have information that another event, has occurred. 

Ex: P(A|B) = probability of event A, if event B occured. 

### **Conditional Probability in Die Toss Example**  

Conditional probability describes the probability of event **A** occurring given that event **B** has already occurred. It is denoted as:

$$
P(A | B) = \frac{P(A)*P(B)}{P(B)}, \quad \text{for } P(B) > 0
$$

#### **Example: Die Toss**  
Let’s define the events:  
#### - **Event A**: Rolling a **6** →  A = {⚅}  
#### - **Event B**: Rolling **at least a 3** → \( B = \{⚂,⚃,⚄,⚅\} \)  

Without additional information, the probability of rolling a 6 is:

$$
P(A) = \frac{1}{6}
$$

Now, suppose we know **B has occurred** (i.e., the roll is among {⚂,⚃,⚄,⚅}). This restricts the possible outcomes to **4 numbers**, making the probability of **A given B**:

$$
P(A | B) = \frac{1}{4}
$$

Since there is **only one favorable outcome (6) out of the four possible values in B**, the probability updates accordingly.

---

### An event A is independent of B if the conditional probability of A given B is the same as the probability of A alone.
$$
    P(A | B) = P(A)
$$

### **Example: Independence of Drawing the Queen of Spades**  

Consider drawing a card from a standard **52-card deck**. Define:  
### - **Event A**: The card is a **spade** ♠️  
### - **Event B**: The card is a **queen** 🃍

To check independence, we compare:

$$
P(A \cap B) = P(\text{Q} \spadesuit) = \frac{1}{52}
$$

with the product of individual probabilities:

$$
P(A) \cdot P(B) = \left(\frac{13}{52}\right) \cdot \left(\frac{4}{52}\right) = \frac{1}{52}
$$

Since P(A|B) = P(A), the events are **independent**.

#### **What if a Card is Removed?**  
If the **2♥ is removed**, the deck now has **51** cards. The probabilities update:

$$
P(A) \cdot P(B) = \frac{13}{51} \cdot \frac{4}{51} \neq P(A \cap B) = \frac{1}{51}
$$

Since the equality no longer holds, **A and B are now dependent**.


#### Bayes’ Rule states that if an event A can occur under **any** of the hypotheses H_i, with a known conditional probability P(A | H_i), and the prior probabilities of hypotheses (H_1, H_2, ..., H_n) are known, then the **posterior probability** of hypothesis H_i given that event A has occurred is:

$$
P(H_i | A) = \frac{P(A | H_i) P(H_i)}{P(A)}
$$

where the **total probability of event \( A \)** is:

$$
P(A) = P(A | H_1) P(H_1) + P(A | H_2) P(H_2) + \dots + P(A | H_n) P(H_n)
$$

This formula allows us to **update** our belief about hypothesis \( H_i \) after observing evidence \( A \). 🚀

## 3.9 Bayesian Networks

 Bayesian networks in which the nodes are events. Many events linked in a causal network form a Bayesian net. Graphically, Bayesian networks are directed acyclic graphs (DAGs) where the nodes represent events and where the directed edges capture their hierarchy and dependence. 

### **Bayesian Network and Conditional Independence**

A sequence of dependencies each node follows a edge to next node:

$$ [A] \rightarrow [B] \rightarrow [C] \rightarrow [D] $$

By conditional independence in Bayesian networks:

$$ P(D | A, B, C) = P(D | C) $$

since **C** is a **parental node** of **D**.

The following example illustrates how **conditioning affects independence**.
