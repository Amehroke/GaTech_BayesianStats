# Q1

Suppose data is generated from the model:

$$
y_i \mid \mu \sim \text{iid} \, N(μ, 1), \; i = 1, \dots, n.
$$

Consider a mixture normal prior:

$$
\mu \sim 0.5N(-1, 1) + 0.5N(1, 1),
$$

that is,

$$
p(\mu) = \frac{0.5}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu+1)^2} + \frac{0.5}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu-1)^2}.
$$

Suppose we have observed y = 1 . Find the posterior distribution of μ.

---

## 1. Start by using Bayes' theorem:

$$
p(\mu \mid y) \propto p(y \mid \mu) p(\mu) = p(y \mid \mu) \Big[ 0.5 \phi(\mu; -1, 1) + 0.5 \phi(\mu; 1, 1) \Big].
$$

Simplify the two components using the following result:

$$
\phi(x; \mu_1, \sigma_1^2) \phi(x; \mu_2, \sigma_2^2) = \phi \left( x; \frac{\mu_1 / \sigma_1^2 + \mu_2 / \sigma_2^2}{1 / \sigma_1^2 + 1 / \sigma_2^2}, \frac{1}{1 / \sigma_1^2 + 1 / \sigma_2^2} \right) \phi(\mu_1 - \mu_2; 0, \sigma_1^2 + \sigma_2^2),
$$

where ϕ(x; μ, σ²) is the density of a normal distribution with mean μ and variance σ².

---

### Step 1: Bayes' Theorem
Using Bayes' theorem:

$$
p(μ | y) ∝ p(y | μ)p(μ) = p(y | μ) [ 0.5 ϕ(μ; -1, 1) + 0.5 ϕ(μ; 1, 1) ].
$$

### Step 2: Simplify the Components
The posterior is a mixture of two terms:
1. ϕ(1; μ, 1)ϕ(μ; -1, 1),
2. ϕ(1; μ, 1)ϕ(μ; 1, 1).

Using the formula:

$$
ϕ(x; μ₁, σ₁²)ϕ(x; μ₂, σ₂²) = ϕ(x; (μ₁ / σ₁² + μ₂ / σ₂²) / (1 / σ₁² + 1 / σ₂²), 1 / (1 / σ₁² + 1 / σ₂²))ϕ(μ₁ - μ₂; 0, σ₁² + σ₂²),
$$

where ϕ(x; μ, σ²) is the normal density.

#### Component 1: ϕ(1; μ, 1)ϕ(μ; -1, 1)
- Variance:
  $$
  σ² = 1 / (1 / 1 + 1 / 1) = 0.5.
  $$
- Mean:
  $$
  μ = (1 / 1 + (-1) / 1) / (1 / 1 + 1 / 1) = 0.
  $$
- Additional factor:
  $$
  ϕ(-1 - 1; 0, 2) = ϕ(-2; 0, 2).
  $$

Thus:

$$
ϕ(1; μ, 1)ϕ(μ; -1, 1) = ϕ(μ; 0, 0.5)ϕ(-2; 0, 2).
$$

#### Component 2: ϕ(1; μ, 1)ϕ(μ; 1, 1)
- Variance:
  $$
  σ² = 1 / (1 / 1 + 1 / 1) = 0.5.
  $$
- Mean:
  $$
  μ = (1 / 1 + 1 / 1) / (1 / 1 + 1 / 1) = 1.
  $$
- Additional factor:
  $$
  ϕ(1 - 1; 0, 2) = ϕ(0; 0, 2).
  $$

Thus:

$$
ϕ(1; μ, 1)ϕ(μ; 1, 1) = ϕ(μ; 1, 0.5)ϕ(0; 0, 2).
$$

### Step 3: Combine the Components
The posterior distribution is:

$$
p(μ | y) ∝ ϕ(-2; 0, 2)ϕ(μ; 0, 0.5) + ϕ(0; 0, 2)ϕ(μ; 1, 0.5).
$$

### Step 4: Normalize the Posterior
The posterior is a mixture of two normal distributions:
- Component 1: ϕ(μ; 0, 0.5) with weight w₁ = ϕ(-2; 0, 2),
- Component 2: ϕ(μ; 1, 0.5) with weight w₂ = ϕ(0; 0, 2).

Normalize the weights:

$$
w₁' = w₁ / (w₁ + w₂), \quad w₂' = w₂ / (w₁ + w₂).
$$

Thus, the posterior is:

$$
p(μ | y) = w₁'ϕ(μ; 0, 0.5) + w₂'ϕ(μ; 1, 0.5).
$$


## 2. The posterior is a mixture of two normal distributions. Identify the following:
   - The two means and variances for the normal distributions.
   - The weights for the two normal components.


### Posterior Distribution of μ

1. **Means and Variances of Components**  
   For each component, the variance \(σ^2\) and mean \(μ\) are computed using the formula:  
   $$
   σ^2 = \frac{1}{\frac{1}{1} + \frac{1}{1}} = 0.5, \quad μ = \frac{\frac{μ_1}{1} + \frac{μ_2}{1}}{\frac{1}{1} + \frac{1}{1}}
   $$

   - **Component 1** (\(μ_1 = -1\)):  
     $$
     μ = \frac{\frac{-1}{1} + \frac{1}{1}}{\frac{1}{1} + \frac{1}{1}} = 0, \quad σ^2 = 0.5.
     $$
     So, the first normal component is:
     $$
     N(0, 0.5).
     $$

   - **Component 2** (\(μ_1 = 1\)):  
     $$
     μ = \frac{\frac{1}{1} + \frac{1}{1}}{\frac{1}{1} + \frac{1}{1}} = 1, \quad σ^2 = 0.5.
     $$
     So, the second normal component is:
     $$
     N(1, 0.5).
     $$

2. **Weights of Components**  
   The unnormalized weights are computed as:  
   $$
   w_1 = 0.5 \cdot \frac{1}{\sqrt{4π}} e^{-1}, \quad w_2 = 0.5 \cdot \frac{1}{\sqrt{4π}}.
   $$

   Normalize the weights:  
   $$
   w_1' = \frac{w_1}{w_1 + w_2}, \quad w_2' = \frac{w_2}{w_1 + w_2}.
   $$

   Substitute:  
   $$
   w_1' = \frac{e^{-1}}{e^{-1} + 1}, \quad w_2' = \frac{1}{e^{-1} + 1}.
   $$

3. **Final Posterior Distribution**  
   The posterior distribution is a mixture of two normal distributions:
   $$
   p(μ | y) = w_1' \cdot N(0, 0.5) + w_2' \cdot N(1, 0.5).
   $$
