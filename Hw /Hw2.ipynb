# Q1

Suppose data is generated from the model:

$$
y_i \mid \mu \sim \text{iid} \, N(μ, 1), \; i = 1, \dots, n.
$$

Consider a mixture normal prior:

$$
\mu \sim 0.5N(-1, 1) + 0.5N(1, 1),
$$

that is,

$$
p(\mu) = \frac{0.5}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu+1)^2} + \frac{0.5}{\sqrt{2\pi}} e^{-\frac{1}{2}(\mu-1)^2}.
$$

Suppose we have observed y = 1 . Find the posterior distribution of μ.

---

## 1. Start by using Bayes' theorem:

$$
p(\mu \mid y) \propto p(y \mid \mu) p(\mu) = p(y \mid \mu) \Big[ 0.5 \phi(\mu; -1, 1) + 0.5 \phi(\mu; 1, 1) \Big].
$$

Simplify the two components using the following result:

$$
\phi(x; \mu_1, \sigma_1^2) \phi(x; \mu_2, \sigma_2^2) = \phi \left( x; \frac{\mu_1 / \sigma_1^2 + \mu_2 / \sigma_2^2}{1 / \sigma_1^2 + 1 / \sigma_2^2}, \frac{1}{1 / \sigma_1^2 + 1 / \sigma_2^2} \right) \phi(\mu_1 - \mu_2; 0, \sigma_1^2 + \sigma_2^2),
$$

where ϕ(x; μ, σ²) is the density of a normal distribution with mean μ and variance σ².

---

### Step 1: Bayes' Theorem
Using Bayes' theorem:

$$
p(μ | y) ∝ p(y | μ)p(μ) = p(y | μ) [ 0.5 ϕ(μ; -1, 1) + 0.5 ϕ(μ; 1, 1) ].
$$

### Step 2: Simplify the Components
The posterior is a mixture of two terms:
1. ϕ(1; μ, 1)ϕ(μ; -1, 1),
2. ϕ(1; μ, 1)ϕ(μ; 1, 1).

Using the formula:

$$
ϕ(x; μ₁, σ₁²)ϕ(x; μ₂, σ₂²) = ϕ(x; (μ₁ / σ₁² + μ₂ / σ₂²) / (1 / σ₁² + 1 / σ₂²), 1 / (1 / σ₁² + 1 / σ₂²))ϕ(μ₁ - μ₂; 0, σ₁² + σ₂²),
$$

where ϕ(x; μ, σ²) is the normal density.

#### Component 1: ϕ(1; μ, 1)ϕ(μ; -1, 1)
- Variance:
  $$
  σ² = 1 / (1 / 1 + 1 / 1) = 0.5.
  $$
- Mean:
  $$
  μ = (1 / 1 + (-1) / 1) / (1 / 1 + 1 / 1) = 0.
  $$
- Additional factor:
  $$
  ϕ(-1 - 1; 0, 2) = ϕ(-2; 0, 2).
  $$

Thus:

$$
ϕ(1; μ, 1)ϕ(μ; -1, 1) = ϕ(μ; 0, 0.5)ϕ(-2; 0, 2).
$$

#### Component 2: ϕ(1; μ, 1)ϕ(μ; 1, 1)
- Variance:
  $$
  σ² = 1 / (1 / 1 + 1 / 1) = 0.5.
  $$
- Mean:
  $$
  μ = (1 / 1 + 1 / 1) / (1 / 1 + 1 / 1) = 1.
  $$
- Additional factor:
  $$
  ϕ(1 - 1; 0, 2) = ϕ(0; 0, 2).
  $$

Thus:

$$
ϕ(1; μ, 1)ϕ(μ; 1, 1) = ϕ(μ; 1, 0.5)ϕ(0; 0, 2).
$$

### Step 3: Combine the Components
The posterior distribution is:

$$
p(μ | y) ∝ ϕ(-2; 0, 2)ϕ(μ; 0, 0.5) + ϕ(0; 0, 2)ϕ(μ; 1, 0.5).
$$

### Step 4: Normalize the Posterior
The posterior is a mixture of two normal distributions:
- Component 1: ϕ(μ; 0, 0.5) with weight w₁ = ϕ(-2; 0, 2),
- Component 2: ϕ(μ; 1, 0.5) with weight w₂ = ϕ(0; 0, 2).

Normalize the weights:

$$
w₁' = w₁ / (w₁ + w₂), \quad w₂' = w₂ / (w₁ + w₂).
$$

Thus, the posterior is:

$$
p(μ | y) = w₁'ϕ(μ; 0, 0.5) + w₂'ϕ(μ; 1, 0.5).
$$

#### Engineering system of type k-out-of-n is operational if at least k out of n components are operational. Otherwise, the system fails. Suppose that a k-out-of-n system consists of n identical and independent elements for which the lifetime has Weibull distribution with parameters r and λ. More precisely, if T is a lifetime of a component,

$$
    P(T \geq t) = e^{-\lambda t^r}, \quad t \geq 0.
$$

#### Time t is in units of months, and consequently, rate parameter λ is in units (month)−1. Parameter r is dimensionless. Assume that n = 10,k = 7, r = 1.3 and λ = 1/20.

## 1. Find the probability that a k-out-of-n system is still operational, when checked at t=6.

#### Step 1: Weibull Survival Function for a Single Component
The survival probability for a single component is given by:
$$
P(T \geq t) = e^{-\lambda t^r}.
$$

Substitute the given values:
- \( \lambda = \frac{1}{20} \),
- \( r = 1.3 \),
- \( t = 6 \).

Compute:
$$
P(T \geq 6) = e^{-\frac{1}{20} \cdot 6^{1.3}}.
$$

First, calculate:
$$
6^{1.3} \approx 9.418.
$$

Then:
$$
P(T \geq 6) = e^{-\frac{1}{20} \cdot 9.418} = e^{-0.4709} \approx 0.6246.
$$

Thus, the survival probability for a single component is:
$$
p \approx 0.6246.
$$

---

#### Step 2: System Operational Probability
The system is operational if at least \( k = 7 \) out of \( n = 10 \) components are operational. The probability is given by:
$$
P(\text{System Operational}) = \sum_{j=7}^{10} \binom{10}{j} p^j (1 - p)^{10 - j}.
$$

Expand the summation:
$$
P(\text{System Operational}) = \binom{10}{7} p^7 (1 - p)^3 + \binom{10}{8} p^8 (1 - p)^2 + \binom{10}{9} p^9 (1 - p)^1 + \binom{10}{10} p^{10}.
$$

---

#### Step 3: Compute Each Term
Using \( p \approx 0.6246 \) and \( 1 - p \approx 0.3754 \):

1. **First term** (\( j = 7 \)):
   $$
   \binom{10}{7} p^7 (1 - p)^3 = \frac{10!}{7! \cdot 3!} (0.6246)^7 (0.3754)^3.
   $$
   Compute:
   $$
   \binom{10}{7} = 120, \quad (0.6246)^7 \approx 0.0589, \quad (0.3754)^3 \approx 0.0529.
   $$
   So:
   $$
   120 \cdot 0.0589 \cdot 0.0529 \approx 0.372.
   $$

2. **Second term** (\( j = 8 \)):
   $$
   \binom{10}{8} p^8 (1 - p)^2 = \frac{10!}{8! \cdot 2!} (0.6246)^8 (0.3754)^2.
   $$
   Compute:
   $$
   \binom{10}{8} = 45, \quad (0.6246)^8 \approx 0.0368, \quad (0.3754)^2 \approx 0.1409.
   $$
   So:
   $$
   45 \cdot 0.0368 \cdot 0.1409 \approx 0.234.
   $$

3. **Third term** (\( j = 9 \)):
   $$
   \binom{10}{9} p^9 (1 - p)^1 = \frac{10!}{9! \cdot 1!} (0.6246)^9 (0.3754).
   $$
   Compute:
   $$
   \binom{10}{9} = 10, \quad (0.6246)^9 \approx 0.0227, \quad (0.3754) \approx 0.3754.
   $$
   So:
   $$
   10 \cdot 0.0227 \cdot 0.3754 \approx 0.085.
   $$

4. **Fourth term** (\( j = 10 \)):
   $$
   \binom{10}{10} p^{10} = \frac{10!}{10! \cdot 0!} (0.6246)^{10}.
   $$
   Compute:
   $$
   \binom{10}{10} = 1, \quad (0.6246)^{10} \approx 0.014.
   $$
   So:
   $$
   1 \cdot 0.014 \approx 0.014.
   $$

---

#### Step 4: Sum the Terms
Add the computed terms:
$$
P(\text{System Operational}) = 0.372 + 0.234 + 0.085 + 0.014 \approx 0.378.
$$

---

### Final Answer
The probability that the system is still operational at time \( t = 6 \) is:
$$
P(\text{System Operational}) \approx 0.378.
$$
```

## 1. What are α and c?

#### Step 1: Likelihood
The likelihood of the data is:
$$
f(u_1, \ldots, u_{35} | \theta) = \prod_{i=1}^{35} \frac{1}{\theta} \cdot 1(\theta > u_i).
$$
This simplifies to:
$$
f(u_1, \ldots, u_{35} | \theta) = \theta^{-35} \cdot 1(\theta > 0.68665).
$$

#### Step 2: Prior
The prior for θ is noninformative (Jeffreys' prior):
$$
\pi(\theta) = \frac{1}{\theta} \cdot 1(\theta > 0).
$$

#### Step 3: Posterior
The posterior is proportional to the product of the likelihood and the prior:
$$
\pi(\theta | \text{data}) \propto f(u_1, \ldots, u_{35} | \theta) \cdot \pi(\theta).
$$
Substituting the likelihood and prior:
$$
\pi(\theta | \text{data}) \propto \theta^{-35} \cdot 1(\theta > 0.68665) \cdot \frac{1}{\theta}.
$$
Simplify:
$$
\pi(\theta | \text{data}) \propto \theta^{-36} \cdot 1(\theta > 0.68665).
$$

#### Step 4: Pareto Distribution
The posterior belongs to the Pareto family:
$$
\pi(\theta | \text{data}) = \frac{\alpha c^\alpha}{\theta^{\alpha+1}} \cdot 1(\theta > c),
$$
where:
- \( α = 35 \),
- \( c = 0.68665 \).

#### Final Answer
1. \( α = 35 \),
2. \( c = 0.68665 \).


## 2. Estimate θ in Bayesian fashion. Then calculate the 95% equi-tailed credible set. Is the true value of parameter (0.7) in the credible set?

#### Step 1: 
The posterior distribution for θ is:
$$
\pi(\theta | \text{data}) = \frac{\alpha c^\alpha}{\theta^{\alpha+1}} \cdot 1(\theta > c),
$$
where:
- α = 35,
- c = 0.68665.

---

#### Step 2: Bayesian Estimate of θ
The Bayesian estimate of θ is the posterior mean:
$$
E[\theta | \text{data}] = \int_{c}^\infty \theta \cdot \pi(\theta | \text{data}) \, d\theta.
$$

Substitute the posterior distribution:
$$
E[\theta | \text{data}] = \int_{c}^\infty \theta \cdot \frac{\alpha c^\alpha}{\theta^{\alpha+1}} \, d\theta.
$$

Simplify:
$$
E[\theta | \text{data}] = \alpha c^\alpha \int_{c}^\infty \theta^{-\alpha} \, d\theta.
$$

Solve the integral:
$$
\int_{c}^\infty \theta^{-\alpha} \, d\theta = \frac{c^{1-\alpha}}{\alpha - 1}.
$$

Substitute back:
$$
E[\theta | \text{data}] = \alpha c^\alpha \cdot \frac{c^{1-\alpha}}{\alpha - 1}.
$$

Simplify:
$$
E[\theta | \text{data}] = \frac{\alpha}{\alpha - 1} \cdot c.
$$

Substitute α = 35 and c = 0.68665:
$$
E[\theta | \text{data}] = \frac{35}{35 - 1} \cdot 0.68665 = \frac{35}{34} \cdot 0.68665 \approx 0.70697.
$$

Thus, the Bayesian estimate of θ is:
$$
\theta \approx 0.707.
$$

#### Step 3: 95% Equi-Tailed Credible Set
The 95% credible set is the interval [θ_L, θ_U] such that:
$$
P(\theta_L \leq \theta \leq \theta_U | \text{data}) = 0.95,
$$
with 2.5% in each tail.

##### Solve for θ_L:
From the cumulative distribution function:
$$
P(\theta \leq \theta_L) = 0.025.
$$

Rearrange:
$$
\left(\frac{c}{\theta_L}\right)^\alpha = 0.975.

Solve for θ_L:
$$

Substitute α = 35 and c = 0.68665:

Compute:

## 3. Plot the posterior PDF, adding marks for the regions bound by the above credible set, along with your point estimate, for each plot.

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha, c = 35, 0.68665

# Define posterior PDF
def posterior_pdf(theta, alpha, c):
    return (alpha * c**alpha) / theta**(alpha + 1) if theta > c else 0

# Generate theta values and posterior probabilities
theta_values = np.linspace(c, 1, 1000)
posterior_values = [posterior_pdf(theta, alpha, c) for theta in theta_values]

# Point estimate and credible set bounds
theta_hat = (alpha / (alpha - 1)) * c
theta_L = c / (0.975 ** (1 / alpha))
theta_U = c / (0.025 ** (1 / alpha))

# Plot posterior PDF
plt.plot(theta_values, posterior_values, label="Posterior PDF", color="blue")
plt.fill_between(theta_values, posterior_values, where=(theta_values >= theta_L) & (theta_values <= theta_U),
                 color="orange", alpha=0.5, label="95% Credible Set")
plt.axvline(theta_hat, color="red", linestyle="--", label=f"Point Estimate (θ̂ = {theta_hat:.3f})")
plt.axvline(theta_L, color="green", linestyle="--", label=f"Lower Bound (θ_L = {theta_L:.3f})")
plt.axvline(theta_U, color="green", linestyle="--", label=f"Upper Bound (θ_U = {theta_U:.3f})")

# Formatting
plt.title("Posterior PDF with 95% Credible Set and Point Estimate")
plt.xlabel("θ")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
````

The plot shows the posterior PDF of θ, highlighting the 95% credible set in orange. It also marks:

- The Bayesian point estimate (θ ≈ 0.707) with a red dashed line.
- The lower bound (θₗ ≈ 0.687) and upper bound (θᵤ ≈ 0.736) of the credible set with green dashed lines.
